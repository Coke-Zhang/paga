# Comparison with DPT

```python
import numpy as np
import pandas as pd
import scanpy.api as sc
sc.settings.verbosity = 1                          # verbosity = 3: errors, warnings, info, hints
sc.settings.set_figure_params(dpi=80)              # low dots per inch yields small inline figures
sc.logging.print_version_and_date()
```

DPT is not able to make a prediction for the continuous and clustering dataset of Figure 1. Moreover, DPT only partitions data into subgoups but does not relate them. Still we compare graph abstraction with it as it is provides a common way of detecting branching points.

## Simple Tree

For the simple tree, DPT is able to detect relatively meaningful partitions of the data. They appear precise when visualized using diffusion maps and less approximate when visualized using tSNE or graph drawing.

```python
adata = sc.datasets.krumsiek11()
sc.tl.tsne(adata)
sc.tl.draw_graph(adata)
sc.tl.dpt(adata, n_branchings=2, n_neighbors=5, knn=False)
```

```python
ax = sc.pl.tsne(adata, color='dpt_groups', title='DPT groups', legend_loc='on data', save=True)
ax = sc.pl.draw_graph(adata, color='dpt_groups', title='DPT groups', save=True)
ax = sc.pl.diffmap(adata, color='dpt_groups', title='DPT groups', components=['1,2', '2,3'], save=True)
```
